# Modelos ARIMA para Series Temporales

## Introducción

ARIMA (AutoRegressive Integrated Moving Average) es una clase de modelos estadísticos para analizar y predecir series temporales. El modelo ARIMA combina tres componentes:

- **AR (AutoRegressive)**: El valor actual depende de valores pasados
- **I (Integrated)**: Diferenciación para hacer la serie estacionaria
- **MA (Moving Average)**: El valor actual depende de errores pasados

Un modelo ARIMA se denota como ARIMA(p,d,q) donde:
- **p**: orden del componente autoregresivo
- **d**: grado de diferenciación
- **q**: orden del componente de media móvil

## 1. Creación de Serie Temporal Sintética

Comenzaremos creando una serie temporal sintética para entender los componentes básicos.

```python
# Configuración de la serie temporal
np.random.seed(42)
n_periods = 365 * 3  # 3 años de datos diarios
dates = pd.date_range(start='2021-01-01', periods=n_periods, freq='D')

# Componentes de la serie temporal
# 1. Tendencia lineal
trend = np.linspace(100, 200, n_periods)

# 2. Estacionalidad anual
seasonal_annual = 20 * np.sin(2 * np.pi * np.arange(n_periods) / 365.25)

# 3. Estacionalidad semanal
seasonal_weekly = 5 * np.sin(2 * np.pi * np.arange(n_periods) / 7)

# 4. Ruido aleatorio
noise = np.random.normal(0, 5, n_periods)

# 5. Componente autorregresivo AR(1)
ar_component = np.zeros(n_periods)
ar_component[0] = noise[0]
for i in range(1, n_periods):
    ar_component[i] = 0.7 * ar_component[i-1] + noise[i]

# Serie temporal completa
y_synthetic = trend + seasonal_annual + seasonal_weekly + ar_component

# Crear DataFrame
df_synthetic = pd.DataFrame({
    'date': dates,
    'value': y_synthetic,
    'trend': trend,
    'seasonal_annual': seasonal_annual,
    'seasonal_weekly': seasonal_weekly,
    'ar_component': ar_component,
    'noise': noise
})
df_synthetic.set_index('date', inplace=True)

print(f"Serie temporal sintética creada con {len(df_synthetic)} observaciones")
print(f"Período: {df_synthetic.index.min()} a {df_synthetic.index.max()}")
df_synthetic.head()
```

## 3. Análisis de Estacionariedad

Para aplicar modelos ARIMA, necesitamos que la serie sea estacionaria. Realizaremos pruebas estadísticas para verificar esto.

```python
def test_stationarity(timeseries, title='Serie Temporal'):
    """
    Realiza pruebas de estacionariedad y muestra resultados gráficos
    """
    print(f'\n=== Análisis de Estacionariedad: {title} ===')
    
    # Estadísticas descriptivas
    print(f'Media: {timeseries.mean():.2f}')
    print(f'Desviación estándar: {timeseries.std():.2f}')
    
    # Test de Dickey-Fuller Aumentado
    print('\n--- Test de Dickey-Fuller Aumentado ---')
    dftest = adfuller(timeseries.dropna(), autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Estadístico ADF', 'p-valor', 
                                           'Lags utilizados', 'Observaciones'])
    for key, value in dftest[4].items():
        dfoutput[f'Valor crítico ({key})'] = value
    
    print(dfoutput)
    
    if dftest[1] <= 0.05:
        print("✓ Rechazamos H0: La serie es ESTACIONARIA")
    else:
        print("✗ No rechazamos H0: La serie es NO ESTACIONARIA")
    
    # Test KPSS
    print('\n--- Test KPSS ---')
    try:
        kpsstest = kpss(timeseries.dropna(), regression='c', nlags="auto")
        kpss_output = pd.Series(kpsstest[0:3], index=['Estadístico KPSS', 'p-valor', 'Lags'])
        for key, value in kpsstest[3].items():
            kpss_output[f'Valor crítico ({key})'] = value
        print(kpss_output)
        
        if kpsstest[1] <= 0.05:
            print("✗ Rechazamos H0: La serie es NO ESTACIONARIA")
        else:
            print("✓ No rechazamos H0: La serie es ESTACIONARIA")
    except Exception as e:
        print(f"Error en test KPSS: {e}")

# Análisis de la serie original
test_stationarity(df_synthetic['value'], 'Serie Original')
```

## 5. Análisis de Autocorrelación (ACF) y Autocorrelación Parcial (PACF)

Estos gráficos nos ayudan a identificar los órdenes p y q del modelo ARIMA.

```python
def plot_acf_pacf(series, title='', lags=40):
    """
    Grafica ACF y PACF para una serie temporal
    """
    fig, axes = plt.subplots(2, 1, figsize=(15, 8))
    
    # ACF
    plot_acf(series.dropna(), lags=lags, ax=axes[0], alpha=0.05)
    axes[0].set_title(f'Autocorrelación (ACF) - {title}')
    axes[0].grid(True, alpha=0.3)
    
    # PACF
    plot_pacf(series.dropna(), lags=lags, ax=axes[1], alpha=0.05, method='ywm')
    axes[1].set_title(f'Autocorrelación Parcial (PACF) - {title}')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Análisis de la serie diferenciada
plot_acf_pacf(df_synthetic['diff1'], 'Primera Diferencia', lags=30)
```

## 6. Identificación Automática de Parámetros ARIMA

Implementaremos una función para encontrar automáticamente los mejores parámetros ARIMA usando criterios de información.

```python
def find_best_arima(series, max_p=5, max_d=2, max_q=5, seasonal=False, m=12):
    """
    Encuentra los mejores parámetros ARIMA usando grid search
    """
    best_aic = float('inf')
    best_params = None
    best_model = None
    results = []
    
    # Generar todas las combinaciones de parámetros
    for p in range(max_p + 1):
        for d in range(max_d + 1):
            for q in range(max_q + 1):
                try:
                    if seasonal:
                        # Para SARIMA
                        for P in range(3):
                            for D in range(2):
                                for Q in range(3):
                                    model = ARIMA(series, order=(p, d, q), 
                                                 seasonal_order=(P, D, Q, m))
                                    fitted = model.fit()
                                    
                                    results.append({
                                        'order': (p, d, q),
                                        'seasonal_order': (P, D, Q, m),
                                        'AIC': fitted.aic,
                                        'BIC': fitted.bic
                                    })
                                    
                                    if fitted.aic < best_aic:
                                        best_aic = fitted.aic
                                        best_params = ((p, d, q), (P, D, Q, m))
                                        best_model = fitted
                    else:
                        # Para ARIMA simple
                        model = ARIMA(series, order=(p, d, q))
                        fitted = model.fit()
                        
                        results.append({
                            'order': (p, d, q),
                            'AIC': fitted.aic,
                            'BIC': fitted.bic
                        })
                        
                        if fitted.aic < best_aic:
                            best_aic = fitted.aic
                            best_params = (p, d, q)
                            best_model = fitted
                            
                except Exception as e:
                    continue
    
    results_df = pd.DataFrame(results).sort_values('AIC')
    
    return best_model, best_params, results_df

# Encontrar el mejor modelo ARIMA
print("Buscando los mejores parámetros ARIMA...")
print("Esto puede tomar unos minutos...")

best_model, best_params, results_df = find_best_arima(
    df_synthetic['value'].dropna(), 
    max_p=3, max_d=2, max_q=3
)

print(f"\nMejor modelo encontrado: ARIMA{best_params}")
print(f"AIC: {best_model.aic:.2f}")
print(f"BIC: {best_model.bic:.2f}")

# Mostrar los 10 mejores modelos
print("\nTop 10 mejores modelos:")
print(results_df.head(10))
```

## 7. Entrenamiento del Modelo ARIMA

```python
# Dividir datos en entrenamiento y prueba
train_size = int(len(df_synthetic) * 0.8)
train_data = df_synthetic['value'][:train_size]
test_data = df_synthetic['value'][train_size:]

print(f"Datos de entrenamiento: {len(train_data)} observaciones")
print(f"Datos de prueba: {len(test_data)} observaciones")

# Entrenar el modelo con los mejores parámetros
model = ARIMA(train_data, order=best_params)
fitted_model = model.fit()

# Resumen del modelo
print("\n=== Resumen del Modelo ARIMA ===")
print(fitted_model.summary())
```

## 8. Diagnóstico del Modelo

Verificaremos si el modelo cumple con los supuestos estadísticos.

```python
# Gráficos de diagnóstico
fitted_model.plot_diagnostics(figsize=(15, 10))
plt.tight_layout()
plt.show()

# Test de Ljung-Box para residuos
residuals = fitted_model.resid
ljung_box_result = acorr_ljungbox(residuals, lags=10, return_df=True)

print("\n=== Test de Ljung-Box para Residuos ===")
print("H0: Los residuos son independientes (ruido blanco)")
print(ljung_box_result)

if (ljung_box_result['lb_pvalue'] > 0.05).all():
    print("\n✓ Los residuos parecen ser ruido blanco (buen modelo)")
else:
    print("\n✗ Los residuos muestran autocorrelación (modelo mejorable)")
```

## 9. Predicciones y Evaluación

```python
# Realizar predicciones
n_forecast = len(test_data)
forecast = fitted_model.forecast(steps=n_forecast)
forecast_ci = fitted_model.get_forecast(steps=n_forecast).conf_int()

# Crear DataFrame con predicciones
forecast_df = pd.DataFrame({
    'forecast': forecast,
    'lower_ci': forecast_ci.iloc[:, 0],
    'upper_ci': forecast_ci.iloc[:, 1]
}, index=test_data.index)

# Calcular métricas de error
mae = mean_absolute_error(test_data, forecast)
mse = mean_squared_error(test_data, forecast)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((test_data - forecast) / test_data)) * 100

print("=== Métricas de Evaluación ===")
print(f"MAE (Error Absoluto Medio): {mae:.2f}")
print(f"MSE (Error Cuadrático Medio): {mse:.2f}")
print(f"RMSE (Raíz del Error Cuadrático Medio): {rmse:.2f}")
print(f"MAPE (Error Porcentual Absoluto Medio): {mape:.2f}%")

# Visualización de predicciones
plt.figure(figsize=(15, 8))

# Datos de entrenamiento
plt.plot(train_data.index, train_data.values, label='Entrenamiento', color='blue', alpha=0.7)

# Datos de prueba (valores reales)
plt.plot(test_data.index, test_data.values, label='Valores Reales', color='green', linewidth=2)

# Predicciones
plt.plot(forecast_df.index, forecast_df['forecast'], label='Predicciones', color='red', linewidth=2)

# Intervalos de confianza
plt.fill_between(forecast_df.index, 
                forecast_df['lower_ci'], 
                forecast_df['upper_ci'], 
                color='red', alpha=0.2, label='Intervalo de Confianza 95%')

plt.title(f'Predicciones ARIMA{best_params}', fontsize=16)
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 12. Comparación de Modelos ARIMA

```python
# Comparar diferentes modelos ARIMA
models_to_compare = [
    (1, 1, 1),
    (2, 1, 2),
    (1, 1, 2),
    (2, 1, 1),
    best_params
]

comparison_results = []

for order in models_to_compare:
    try:
        # Entrenar modelo
        model = ARIMA(train_data, order=order)
        fitted = model.fit()
        
        # Predicciones
        pred = fitted.forecast(steps=len(test_data))
        
        # Métricas
        mae = mean_absolute_error(test_data, pred)
        mse = mean_squared_error(test_data, pred)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs((test_data - pred) / test_data)) * 100
        
        comparison_results.append({
            'Modelo': f'ARIMA{order}',
            'AIC': fitted.aic,
            'BIC': fitted.bic,
            'MAE': mae,
            'RMSE': rmse,
            'MAPE': mape
        })
        
    except Exception as e:
        print(f"Error con modelo ARIMA{order}: {e}")

# Crear DataFrame de comparación
comparison_df = pd.DataFrame(comparison_results)
comparison_df = comparison_df.round(2)

print("\n=== Comparación de Modelos ARIMA ===")
print(comparison_df.to_string(index=False))

# Encontrar el mejor modelo por cada métrica
print("\n=== Mejores Modelos por Métrica ===")
for metric in ['AIC', 'BIC', 'MAE', 'RMSE', 'MAPE']:
    best_idx = comparison_df[metric].idxmin()
    best_model = comparison_df.loc[best_idx, 'Modelo']
    best_value = comparison_df.loc[best_idx, metric]
    print(f"{metric}: {best_model} ({best_value})")
```

## 13. Funciones Auxiliares para Análisis ARIMA

```python
class ARIMAAnalyzer:
    """
    Clase para facilitar el análisis completo de modelos ARIMA
    """
    
    def __init__(self, data, date_column=None):
        if isinstance(data, pd.DataFrame):
            if date_column:
                data = data.set_index(date_column)
            self.data = data.iloc[:, 0]  # Tomar la primera columna
        else:
            self.data = data
        
        self.model = None
        self.fitted_model = None
        
    def check_stationarity(self, series=None):
        """Verificar estacionariedad"""
        if series is None:
            series = self.data
        
        test_stationarity(series)
        
    def difference_series(self, d=1):
        """Aplicar diferenciación"""
        differenced = self.data.copy()
        for i in range(d):
            differenced = differenced.diff().dropna()
        return differenced
    
    def plot_decomposition(self, model='additive', period=None):
        """Descomposición de serie temporal"""
        if period is None:
            # Estimar período automáticamente
            period = min(len(self.data) // 2, 365 if 'D' in str(self.data.index.freq) else 12)
        
        decomposition = seasonal_decompose(self.data.dropna(), model=model, period=period)
        
        fig, axes = plt.subplots(4, 1, figsize=(15, 12))
        
        decomposition.observed.plot(ax=axes[0], title='Serie Original')
        decomposition.trend.plot(ax=axes[1], title='Tendencia')
        decomposition.seasonal.plot(ax=axes[2], title='Estacionalidad')
        decomposition.resid.plot(ax=axes[3], title='Residuos')
        
        for ax in axes:
            ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        return decomposition
    
    def auto_arima(self, max_p=5, max_d=2, max_q=5):
        """Encontrar automáticamente los mejores parámetros"""
        self.fitted_model, self.best_params, self.results_df = find_best_arima(
            self.data, max_p, max_d, max_q
        )
        return self.fitted_model, self.best_params
    
    def fit_arima(self, order):
        """Ajustar modelo ARIMA con parámetros específicos"""
        self.model = ARIMA(self.data, order=order)
        self.fitted_model = self.model.fit()
        return self.fitted_model
    
    def forecast(self, steps=30, alpha=0.05):
        """Realizar predicciones"""
        if self.fitted_model is None:
            raise ValueError("Modelo no ajustado. Use fit_arima() o auto_arima() primero.")
        
        forecast = self.fitted_model.forecast(steps=steps)
        forecast_ci = self.fitted_model.get_forecast(steps=steps).conf_int(alpha=alpha)
        
        return forecast, forecast_ci
    
    def plot_forecast(self, steps=30, historical_periods=100):
        """Visualizar predicciones"""
        forecast, forecast_ci = self.forecast(steps)
        
        # Crear fechas futuras
        last_date = self.data.index[-1]
        if hasattr(self.data.index, 'freq') and self.data.index.freq:
            future_dates = pd.date_range(start=last_date + self.data.index.freq, 
                                       periods=steps, freq=self.data.index.freq)
        else:
            # Inferir frecuencia
            freq = pd.infer_freq(self.data.index)
            if freq is None:
                freq = 'D'  # Default a diario
            future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), 
                                       periods=steps, freq=freq)
        
        plt.figure(figsize=(15, 8))
        
        # Datos históricos
        historical_data = self.data[-historical_periods:]
        plt.plot(historical_data.index, historical_data.values, 
                label='Datos Históricos', color='blue', linewidth=2)
        
        # Predicciones
        plt.plot(future_dates, forecast, 
                label='Predicciones', color='red', linewidth=2, linestyle='--')
        
        # Intervalo de confianza
        plt.fill_between(future_dates, 
                        forecast_ci.iloc[:, 0], 
                        forecast_ci.iloc[:, 1], 
                        color='red', alpha=0.2, label='IC 95%')
        
        plt.title('Predicciones ARIMA', fontsize=16)
        plt.xlabel('Fecha')
        plt.ylabel('Valor')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

# Ejemplo de uso de la clase
print("\n=== Ejemplo de uso de ARIMAAnalyzer ===")

# Crear analizador
analyzer = ARIMAAnalyzer(df_synthetic['value'])

# Descomposición
decomp = analyzer.plot_decomposition(period=365)

# Ajustar modelo automáticamente
model, params = analyzer.auto_arima(max_p=2, max_d=2, max_q=2)

print(f"Mejor modelo automático: ARIMA{params}")

# Visualizar predicciones
analyzer.plot_forecast(steps=60, historical_periods=200)
```

## 14. Resumen y Conclusiones

### Pasos del Análisis ARIMA:

1. **Análisis Exploratorio**: Visualización de la serie temporal
2. **Verificación de Estacionariedad**: Tests ADF y KPSS
3. **Diferenciación**: Si es necesario para lograr estacionariedad
4. **Identificación de Parámetros**: Análisis ACF y PACF
5. **Estimación del Modelo**: Ajuste de parámetros
6. **Diagnóstico**: Verificación de supuestos del modelo
7. **Predicción**: Forecasting y evaluación

### Interpretación de Parámetros:

- **p (AR)**: Número de observaciones pasadas que influyen en el valor actual
- **d (I)**: Número de diferenciaciones necesarias para estacionariedad
- **q (MA)**: Número de errores pasados que influyen en el valor actual

### Criterios de Selección:

- **AIC/BIC**: Menores valores indican mejor modelo
- **Residuos**: Deben ser ruido blanco (independientes)
- **Precisión**: MAE, RMSE, MAPE en datos de prueba

### Limitaciones de ARIMA:

- Asume linealidad
- Requiere estacionariedad
- No maneja bien cambios estructurales
- Limitado con series con múltiples estacionalidades

### Extensiones:

- **SARIMA**: Para estacionalidad
- **ARIMAX**: Con variables exógenas
- **Vector ARIMA**: Para múltiples series
